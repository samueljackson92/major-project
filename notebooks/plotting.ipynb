```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from pandas.tools import plotting
import pandas as pd
from sklearn import preprocessing

df = pd.DataFrame.from_csv('./results/2015-03-05-results-features.csv')
columns = ['min_radius', 'max_radius', 'avg_radius', 'std_radius', 'avg_std_intensity', 'avg_avg_intensity']
df_temp = df[columns]
data = preprocessing.normalize(df_temp)
data = pd.DataFrame(data, columns=columns)

data.index = df.index
data['class'] = df['class']
data = data.sort('class')
plotting.radviz(data, 'class', colormap=cm.Spectral)
plt.show()
```
